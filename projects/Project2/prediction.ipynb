### This model has been built using Support Vector Machine

**importing dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**reading the csv file**

```python
diabetes_data = pd.read_csv('./diabetes.csv')
```

```python
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

**Here,**
* 0 -> Non Diabetic
* 1 -> Diabetic

```python
X = diabetes_data.drop(columns='Outcome', axis=1)
Y = diabetes_data['Outcome']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

**Standardizing the data**

```python
standardized_data = scaler.transform(X)
```

```python
x = standardized_data
```

**splitting the data for training and testing**

```python
x_train, x_test, Y_train, Y_test = train_test_split(x,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
classifier = svm.SVC(kernel='linear')
```

**training the data**

```python
classifier.fit(x_train,Y_train)
```

**training data accuracy**

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,Y_train)
print(f"Accuracy score of training data: {training_data_accuracy}")
```

**test data accuracy**

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,Y_test)
print(f"Accuracy score of test data: {training_data_accuracy}")
```

**testing the model**

```python
input_data = (4,146,92,0,0,31.2,0.539,61)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
standardized_reshaped_data = scaler.transform(input_data_reshaped)
```

```python
prediction = classifier.predict(standardized_reshaped_data)
if(prediction[0]==1):
    print("Diabetic")
else:
    print("Not Diabetic")
```
