### This model has been built using Logistic Regression

**importing the dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**data collection and processing**

```python
credit_card_data = pd.read_csv('./creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data['Class'].value_counts()
```

* 0 -> normal transaction
* 1 -> fraud transaction

**separating the data for analysis**

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

**statistical measures of data**

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

**compare the values for both transaction**

```python
credit_card_data.groupby('Class').mean()
```

**under sampling**

```python
legit_sample = legit.sample(n=492)
```

**concatenating two data**

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```

**splitting data into features & targets**

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

**split the data into training and testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

**model training**

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train,Y_train)
```

**evaluation**

```python
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(f"accuracy score on training data {train_data_accuracy}")
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print(f"accuracy score on test data {test_data_accuracy}")
```
